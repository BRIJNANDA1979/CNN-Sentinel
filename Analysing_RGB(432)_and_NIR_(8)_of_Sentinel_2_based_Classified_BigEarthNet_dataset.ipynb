```python
# Add batch of RGB band(432) and NIR band(8)tif files to files list
path = '/content/drive/MyDrive/Big'
dirs = os.listdir(path)
del dirs[-1]
band_code = [2,3,4,8] #band files to be extracted
files_batch=[]
for i in band_code:
  for j in dirs:
    s=''
    s = s + path + '/' + str(j) +  '/' +'*0' + str(i) + '.tif'
    #print(s)
    temp = (glob.glob(os.path.join(s)))
    files_batch.append(temp[0])
#print(files_batch)
print(len(files_batch))
```

```python
# Add min/max values of bands(234 & 8) to min/max list
min=[]
max=[]
mean=[]

for i in files_batch:
    array = rxr.open_rasterio(i)
    min.append(np.nanmin(array))
    max.append(np.nanmax(array))
    mean.append(np.nanmean(array))
print(len(min))
print(len(max))
print(len(mean))
```

```python
#Split 24 elements in group of 4 : one for eacvh of bands (234 & 8)
temp=np.array_split(mean, 4)
mean_band2=list(temp[0])
mean_band3=list(temp[1])
mean_band4=list(temp[2])
mean_band8=list(temp[3])
#print(t.sum()/6)
```

```python
# Add filename and min/max lists to Dataframe
df=pd.DataFrame()
#df['filename'] = files_batch
df['mean_band2'] = mean_band2
df['mean_band3'] = mean_band3
df['mean_band4'] = mean_band4
df['mean_band8'] = mean_band8
print(df)
```
